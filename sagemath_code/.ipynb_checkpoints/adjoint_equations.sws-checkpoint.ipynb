```python
var('s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c, w_f, u_l_t, v_h_t');
var('psi_s_f, psi_l_f, psi_i_f, psi_s_c, psi_l_c, psi_i_c_l, psi_i_c_h, psi_t_c');
var('lambda_f, beta_f, beta_f_tilde, n_c_inf, mu_f, k_f');
var('lambda_c, beta_c, beta_c_tilde, n_c_inf, mu_c, k_c, theta, rho');
var('a_l_f, a_i_f, a_l_c, a_i_c_l, a_i_c_h, b_f')
#
rhs_s_f(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c) = lambda_f - beta_f / n_c_inf * (s_f * i_c_l) \
                                - beta_f_tilde / n_c_inf * (s_f * i_c_h) - (mu_f + w_f) * s_f;
rhs_l_f(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c) = beta_f / n_c_inf * (s_f * i_c_l) \
                                + beta_f_tilde / n_c_inf * (s_f * i_c_h) - (k_f + mu_f + w_f) * l_f;
rhs_i_f(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c) = k_f * l_f - (mu_f +w_f) *i_f;
#
rhs_s_c(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c) = lambda_c - beta_c / n_c_inf * (s_c * i_f) - mu_c * s_c;
rhs_l_c(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c) = beta_c / n_c_inf * (s_c * i_f) - (mu_c + k_c) * l_c;
#
rhs_i_c_l(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c) = theta * k_c * l_c - (mu_c + u_l_t) * i_c_l \
                                                        - beta_c_tilde / n_c_inf * i_c_l * i_f;
rhs_i_c_h(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c) = (1.0 - theta) * k_c * l_c - (mu_c + v_h_t) * i_c_h\
                                                        + beta_c_tilde / n_c_inf * i_c_l * i_f;

rhs_t_c(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c) = v_h_t * i_c_h - (rho + mu_c) * t_c;
rhs = vector(
        (
            rhs_s_f(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c), \
            rhs_l_f(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c), \
            rhs_i_f(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c), \
            rhs_s_c(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c), \
            rhs_l_c(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c), \
            rhs_i_c_l(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c), \
            rhs_i_c_h(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c), \
            rhs_t_c(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c)
        )
    )
#
g_cost(s_f, l_f, i_f, s_c, l_c, i_c_l, i_c_h, t_c) = \
    a_l_f * l_f + a_i_f * i_f + a_l_c * l_c + a_i_c_l * i_c_l + a_i_c_h * i_c_h + b_f * (w_f ^ 2);
psi = vector((psi_s_f, psi_l_f, psi_i_f, psi_s_c, psi_l_c, psi_i_c_l, psi_i_c_h, psi_t_c));
H = rhs.inner_product(psi)
latex(H)
```

```python
rhs(s_f=1, l_f=0, i_f=0, s_c=0, l_c=0, i_c_l=0, i_c_h=0, t_c=0);
g_cost(1, 1, 1, 1, 1, 1, 1)
```

```python
# Adjoint equations 
rhs_psi_s_f = H.diff(s_f)
rhs_psi_l_f = H.diff(l_f)
rhs_psi_i_f = H.diff(i_f)
rhs_psi_s_c = H.diff(s_c)
rhs_psi_l_c = H.diff(l_c)
rhs_psi_i_c_l = H.diff(i_c_l)
rhs_psi_i_c_h = H.diff(i_c_h)
rhs_psi_t_c = H.diff(t_c)
#
latex(rhs_psi_l_c )
```
