```python
var('s_f, l_f, i_f, s_c, l_c, i_c, t_c, w_f, u_t, v_t');
var('psi_s_f, psi_l_f, psi_i_f, psi_s_c, psi_l_c, psi_i_c, psi_t_c')
var('lambda_f, beta_f, n_c_inf, mu_f, k_f');
var('lambda_c, beta_c, n_c_inf, mu_c, k_c, delta, rho');
var('a_l_f, a_i_f, a_l_c, a_i_c, b_f')
#
rhs_s_f(s_f, l_f, i_f, s_c, l_c, i_c, t_c) = lambda_f - beta_f / n_c_inf * (s_f * i_c) - (mu_f + w_f) * s_f;
rhs_l_f(s_f, l_f, i_f, s_c, l_c, i_c, t_c) = beta_f / n_c_inf * (s_f * i_c) - (k_f + mu_f + w_f) * l_f;
rhs_i_f(s_f, l_f, i_f, s_c, l_c, i_c, t_c) = k_f * l_f - (mu_f +w_f) *i_f;
#
rhs_s_c(s_f, l_f, i_f, s_c, l_c, i_c, t_c) = lambda_c - beta_c / n_c_inf * (s_c * i_f) - mu_c * s_c;
rhs_l_c(s_f, l_f, i_f, s_c, l_c, i_c, t_c) = beta_c / n_c_inf * (s_c * i_f) - (mu_c + k_c) * l_c;
rhs_i_c(s_f, l_f, i_f, s_c, l_c, i_c, t_c) = k_c * l_c - mu_c + delta * i_c;

rhs_t_c(s_f, l_f, i_f, s_c, l_c, i_c, t_c) = delta * u_t * i_c - (rho + mu_c) * t_c;
rhs = vector(
        (
            rhs_s_f(s_f, l_f, i_f, s_c, l_c, i_c, t_c), \
            rhs_l_f(s_f, l_f, i_f, s_c, l_c, i_c, t_c), \
            rhs_i_f(s_f, l_f, i_f, s_c, l_c, i_c, t_c), \
            rhs_s_c(s_f, l_f, i_f, s_c, l_c, i_c, t_c), \
            rhs_l_c(s_f, l_f, i_f, s_c, l_c, i_c, t_c), \
            rhs_i_c(s_f, l_f, i_f, s_c, l_c, i_c, t_c), \
            rhs_t_c(s_f, l_f, i_f, s_c, l_c, i_c, t_c)
        )
    )
#
g_cost(s_f, l_f, i_f, s_c, l_c, i_c, w_f) = \
    a_l_f * l_f + a_i_f * i_f + a_l_c * l_c + a_i_c * i_c + b_f * (w_f ^ 2);
psi = vector((psi_s_f, psi_l_f, psi_i_f, psi_s_c, psi_l_c, psi_i_c, psi_t_c));
H = rhs.inner_product(psi)
```

```python
rhs(s_f=1, l_f=0, i_f=0, s_c=0, l_c=0, i_c=0);
g_cost(1, 1, 1, 1, 1, 1, 1)
```

```python
# Adjoint equations 
rhs_psi_s_f = H.diff(s_f)
rhs_psi_l_f = H.diff(l_f)
rhs_psi_i_f = H.diff(i_f)
rhs_psi_s_c = H.diff(s_c)
rhs_psi_l_c = H.diff(l_c)
rhs_psi_i_c = H.diff(i_c)
rhs_psi_t_c = H.diff(t_c)
```

```python
rhs_psi = vector(
            (
                rhs_psi_s_f, \
                rhs_psi_l_f, \
                rhs_psi_i_f, \
                rhs_psi_s_c, \
                rhs_psi_l_c, \
                rhs_psi_i_c, \
                rhs_psi_t_c  \
            )
        )
```

The regarding Hamiltonian results
\begin{equation}
    H(x, \psi)=
    {
        \left(
            \delta i_{c} + k_{c} l_{c} - \mu_{c}
        \right)
    }
    \psi_{i_{c}} 
    + 
    {
        \left(
            k_{f} l_{f} - i_{f}
            {
                \left(
                    \mu_{f} + w_{f}
                \right)
            }
        \right)
    } 
    \psi_{i_{f}} 
    - {
        \left(
            {
                \left(
                    k_{c} 
                    + 
                    \mu_{c}
        \right)
     } l_{c} - 
     \frac{\beta_{c} i_{f} s_{c}}{n_{c_{\mathit{inf}}}}\right)} \psi_{l_{c}} -
     {
         \left(
             {
                 \left(
                     k_{f} + \mu_{f} + w_{f}
                 \right)
             } l_{f} - 
             \frac{\beta_{f} i_{c} s_{f}}{n_{c_{\mathit{inf}}}}
         \right)
     } 
     \psi_{l_{f}} - 
     {
         \left(
             \mu_{c} s_{c} 
             + 
             \frac{\beta_{c} i_{f} s_{c}}{n_{c_{\mathit{inf}}}} - \lambda_{c}
         \right)
     } \psi_{s_{c}} - 
     {
         \left(
             {
                 \left(
                     \mu_{f} + w_{f}
                 \right)
             } s_{f} + 
          \frac{\beta_{f} i_{c} s_{f}}{n_{c_{\mathit{inf}}}} 
          - \lambda_{f}\right)} \psi_{s_{f}} 
          + 
          {
              \left(
                  \delta i_{c} u_{t} - 
                  {
                      \left(
                          \mu_{c} 
                          + \rho
                      \right)
                  } t_{c}
              \right)
          } \psi_{t_{c}} .    
\end{equation}

We have to solve backwards the adjoint ODE

\begin{equation}
    \begin{aligned}
        \dot
        \psi_{s_f} &=
            \frac{\beta_{f} i_{c} \psi_{l_{f}}}{n_{c_{\mathit{inf}}}} 
            - {
                \left(
                    \mu_{f} 
                    + \frac{\beta_{f} i_{c}}{n_{c_{\mathit{inf}}}} + w_{f}
                \right)
              } \psi_{s_{f}}
        \\
        \dot
        \psi_{l_f} &=
              k_{f} \psi_{i_{f}} 
              - {
                  \left(
                      k_{f} 
                      + \mu_{f} 
                      + w_{f}
                  \right)
                } 
                \psi_{l_{f}}
          \\
          \dot
          \psi_{i_f} &=
              -{
                    \left(
                        \mu_{f} + 
                        w_{f}
                    \right)
                } \psi_{i_{f}} 
                 + 
                 \frac{\beta_{c} \psi_{l_{c}} s_{c}}{n_{c_{\mathit{inf}}}} 
                 - \frac{\beta_{c} \psi_{s_{c}} s_{c}}{n_{c_{\mathit{inf}}}}
         \\
         \dot
         \psi_{s_c} &=
             \frac{\beta_{c} i_{f} \psi_{l_{c}}}{n_{c_{\mathit{inf}}}} 
             - {
                     \left(
                         \mu_{c}
                         + 
                         \frac{\beta_{c} i_{f}}{n_{c_{\mathit{inf}}}}
                     \right)
               } 
               \psi_{s_{c}}
          \\
          \dot
          \psi_{l_c} &=
              k_{c} \psi_{i_{c}} - 
                   {
                       \left(k_{c} 
                       + \mu_{c}
                       \right)
                   } \psi_{l_{c}}
          \\
          \dot
          \psi_{i_c} &=
                   \delta \psi_{t_{c}} u_{t} 
                       + \delta \psi_{i_{c}} 
                       + \frac{\beta_{f} \psi_{l_{f}} s_{f}}{n_{c_{\mathit{inf}}}} 
                       - \frac{\beta_{f} \psi_{s_{f}} s_{f}}{n_{c_{\mathit{inf}}}}
           \\
           \dot
           \psi_{t_c} &=
               -{\left(\mu_{c} + \rho\right)} \psi_{t_{c}}
    \end{aligned}
\end{equation}
